```python
import pandas as pd
import numpy as np 
import matplotlib.pyplot as plt

#DataFrame with students
data = {
    'student_name': ['Nastya', 'Vladimir', 'Oleg', 'Marina', 'Nikita'],
    'attendance': [85, 60, 40, 95, 100],  
    'mark': [75, 65, 60, 80, 90]         
}

df = pd.DataFrame(data)

print(df)

plt.figure(figsize=(8, 5))
plt.scatter(df['attendance'], df['mark'], color='blue', s=100)

plt.title('Correlation between Attendance and Mark')
plt.xlabel('Attendance (%)')
plt.ylabel('Marks (out of 100)')

m, b = np.polyfit(df['attendance'], df['mark'], 1)
plt.plot(df['attendance'], m * df['attendance'] + b, color='red')

plt.grid()
plt.show()
```
